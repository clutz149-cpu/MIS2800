```python
%%sql
/* 1. Create the table with STRICT types */
CREATE TABLE shoes (
    shoe_id INTEGER PRIMARY KEY,
    shoe_name TEXT,
    cost REAL
) STRICT;

INSERT INTO shoes (shoe_name, cost) VALUES ('Air Jordan 1', 170.00);
INSERT INTO shoes (shoe_name, cost) VALUES ('Nike Dunk', 200.00);
INSERT INTO shoes (shoe_name, cost) VALUES ('Nike Blazer', 110.00);

```

```python
%%sql

SELECT * FROM shoes WHERE cost < 200;
```

```python
%%sql

UPDATE shoes
SET cost = 185.00
WHERE shoe_name = 'Air Jordan 1';

DELETE FROM shoes
WHERE shoe_id = 1;

SELECT * FROM shoes
```
